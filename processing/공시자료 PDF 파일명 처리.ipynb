```python
import os
import shutil

pdf_directory = os.path.join(os.getcwd(), "셀트리온 공시자료")

pdf_files = [os.path.splitext(f)[0] for f in os.listdir(pdf_directory) if f.lower().endswith('.pdf')]

pdf_data_directory = os.path.join(os.getcwd(), "pdf_data")
if not os.path.exists(pdf_data_directory):
    os.makedirs(pdf_data_directory)

print(f"총 {len(pdf_files)}개의 PDF 파일을 찾았습니다.")
print("\n파일 목록:")
for i, file in enumerate(pdf_files, 1):
    print(f"{i}. {file}")
    original_path = os.path.join(pdf_directory, f"{file}.pdf")
    new_filename = f"celltrion_{i}.pdf"
    new_path = os.path.join(pdf_data_directory, new_filename)
    print(original_path, new_path)
    shutil.copy2(original_path, new_path)

print("\n파일 리스트 변수:")
print(pdf_files)
```
